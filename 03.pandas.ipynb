## 1. DataFrame과 list, dictionary, numpay의 ndarray 상호 변환

```python
# 1차원 데이터 -> Dataframe
col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

#list를 이용해 df 생성
df_list1 = pd.DataFrame(list1, columns= col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

#ndarray를 이용해 df 생성
df_list2 = pd.DataFrame(array1, columns= col_name1)
print('1차원 ndarray를 만든 DataFrame:\n', df_list2)
```

```python
# 2차원 데이터 -> Dataframe
col_name2 = ['col1','col2','col3']
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print('array2 shape:',array2.shape)

#list를 이용해 df 생성
df_list2 = pd.DataFrame(list2,columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)

#ndarray를 이용해 df 생성
df_list3 = pd.DataFrame(array2, columns= col_name2)
print('2차원 ndarray를 만든 DataFrame:\n', df_list3)
```

```python
# dictionary -> Dataframe
# key는 문자열 컬럼, value는 리스트 or ndarray 칼럼 데이터로 매핑 
dict = {'col':[1,11], 'col2':[2,22], 'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print(df_dict)
```

```python
# DataFrame -> ndarray : values 
array3 = df_dict.values
print('df_dict.values 타입', type(array3), 'df_dict.values shape', array3.shape)
print(array3)
```

```python
# DataFrame -> list : values.tolist()
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입', type(list3))
print(list3)

# DataFrame -> dictionary
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict 타입', type(dict3))
print(dict3)
```

### 5-1. iloc 연산자

- 위치 기반 인덱싱 (행, 열 위치 입력)

```python
data = {'Name' : ['Jenny','Jisoo','Lisa','Rose','Karina'] ,
        'Year' : [2012,2011,2010,2009,2006],
        'Gender' : ['Male','Male','Female','Female','Female'] 
        }

data_df = pd.DataFrame(data, index= ['one','two','three','four','five'])
data_df
```

```python
data_df.iloc[0,0]
```

```python
data_df.iloc[3,2]
```

```python
# 행 위치에 인덱스 값인 'one'을 입력하여 오류 발생 
data_df.iloc['one',2]
```

```python
print('마지막 컬럼 데이터\n', data_df.iloc[:,-1])
print('맨 마지막 컬럼을 제외한 모든 데이터\n', data_df.iloc[:,:-1])
```

### 5-2. loc 연산자

- label 기반 데이터 추출

```python
data_df.loc['one','Name']
```

```python
print('위치기반 iloc slicing \n', data_df.iloc[0:1, 0],'\n')
print('명칭기반 loc slicing \n', data_df.loc['one':'two', 'Name'],'\n')
```
